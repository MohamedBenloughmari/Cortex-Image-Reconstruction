```python

import torch
def Linear_Gaussian(n):
    # For a single sample from N(0, Sigma) where Sigma is 5x5
    Sigma=torch.eye(5)
    X1 = torch.distributions.MultivariateNormal(
        loc=torch.zeros(5), 
        covariance_matrix=Sigma
        ).sample()
    X=[X1]
    Y=[]
    A=torch.rand(size=(5,5))
    B=torch.rand(size=(5,5))
    C=torch.rand(size=(5,5))
    D=torch.rand(size=(5,5))

    for _ in range(n):
        X.append(A@X[-1]+B@torch.randn(5))
        Y.append(C@X[-1]+D@torch.randn(5))
    return torch.stack(X),torch.stack(Y)


X,Y=Linear_Gaussian(5)
print(X)
print(Y)
```
